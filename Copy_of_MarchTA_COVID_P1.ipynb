```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
target_en=lb.fit_transform(target)
print(target_en.tolist())
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(data_image,target_en,test_size=0.20,random_state=42)
```

```python
import torch
import torchvision
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
import torch.nn.functional as F
from torch.utils.data import Dataset,DataLoader
transform=transforms.Compose([transforms.Resize((100,100)),transforms.ToTensor()])
class Data(Dataset):
  def __init__(self,X,Y):
    self.X=X
    self.Y=Y
    self.len=len(self.X)

  def __getitem__(self,index):
    X=self.X[index]
    X=X.convert("RGB")
    return transform(X),self.Y[index]

  def __len__(self):
    return self.len
```

```python
plt.imshow(x_train[0].convert("RGB"))
```

```python
train=Data(x_train,y_train)
test=Data(x_test,y_test)
batch_size=16
trainloader=DataLoader(train,batch_size=batch_size,shuffle=True)
testloader=DataLoader(test,batch_size=batch_size,shuffle=False)
```

```python
import torchvision
from torchvision import models
#model=models.resnet50(pretrained=True)
model=models.mobilenet_v2(pretrained=True)
```

```python
import torch.optim as optim
from tqdm import tqdm
criterion=nn.CrossEntropyLoss()
optimizer=optim.Adam(model.parameters(),lr=0.0001)
num_epochs=1

for epoch in tqdm (range(num_epochs)):
  for image,label in trainloader:
    optimizer.zero_grad()
    outputs=model(image)
    loss=criterion(outputs,label)
    loss.backward()
    optimizer.step()
  print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
```

```python
total=0                                         # total number of samples processed
correct=0                                        # number of correctly classified samples
itest=50                                          # number of iterations
for images,labels in testloader:
      optimizer.zero_grad()
      output = model(images)        # model is applied to batch of images fetched by iterations to obtain predictions
      _,out=torch.max(output,1)                   #out is predicted class label
      correct+=torch.sum(out==labels.data)        # number of correct predictions
      total+=len(labels)
      optimizer.step()
accuracy=correct/total
print("Correct:",correct,"\nTotal:",total,"\nAccuracy:",accuracy)
```

```python
import numpy as np
import cv2
def add_scatter_noise(image, intensity=0.10):
    w, h= image.size
    c=3
    noise = np.zeros((h, w, c), dtype=np.float32)
    cv2.randn(noise, 0, intensity * 255)  # Generate random Gaussian noise
    noisy_image = np.clip(image + noise, 0, 255).astype(np.uint8)
    return noisy_image
```

```python
noise_image_path=[]
for i in range(0,50):
  test_image=x_test[i].convert("RGB")
  noisy_image = add_scatter_noise(test_image)
  noise_image_path.append("/content/drive/MyDrive/Colab-Notebooks/noise_image_"+str(i)+".jpg")
  cv2.imwrite("/content/drive/MyDrive/Colab-Notebooks/noise_image_"+str(i)+".jpg", noisy_image)
```
